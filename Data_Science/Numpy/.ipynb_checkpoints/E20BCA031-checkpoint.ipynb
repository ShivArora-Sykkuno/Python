### Creation of Data Frame.

```python
data = pd.DataFrame({"":["June", "Robert", "Lily","David", "Namn", "Vinay", "Param", "Aman", "Lily","Param"],
                     "Apple": [3,2,0,1,13,None,7,6,0,7], 
                     "Mango":[0,2,4,5,1,6,8,None,4,8],"Orange":[2,3,5,7,9,1,0,11,5,0]})
data = data.set_index("")
data
```

```python
Class: DIE, LIVE
 2. AGE: 10, 20, 30, 40, 50, 60, 70, 80
 3. SEX: male, female
 4. STEROID: no, yes
 5. ANTIVIRALS: no, yes
 6. FATIGUE: no, yes
 7. MALAISE: no, yes
 8. ANOREXIA: no, yes
 9. LIVER BIG: no, yes
 10. LIVER FIRM: no, yes
 11. SPLEEN PALPABLE: no, yes
 12. SPIDERS: no, yes
 13. ASCITES: no, yes
 14. VARICES: no, yes
 15. BILIRUBIN: 0.39, 0.80
    16. ALK PHOSPHATE: 33, 80, 120, 160, 200, 250
 17. SGOT: 13, 100, 200, 300, 400, 500,
 18. ALBUMIN: 2.1, 3.0, 3.8, 4.5, 5.0, 6.0
 19. PROTIME: 10, 20, 30, 40, 50, 60, 70, 80, 90
 20. HISTOLOGY: no, yes
```

```python
health = pd.read_csv("hepatitis.data",
                     names=["Class", "Age", "Sex", "Steriod","Antivirals",
                            "Fatigue", "Malaise", "Anorexia", "Liver Bag", "Liver Firm", 
                            "Spleen Palpable", "Spiders", "Ascites", "Varices", "Bilirubin",
                            "Alk Phosphate", "Sgot", "Albumin", "Protime", "Histology"],
                     header=None, )
health
```

```python
health.info, health.shape
```
