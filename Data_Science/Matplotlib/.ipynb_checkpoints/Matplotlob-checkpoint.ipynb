# Anatomy of a MatplotLib Plot

### Matplotlib Example Workflow

```python
# 0. import the Matplot lib library and get it ready forr plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Pepare the Data
x = [1,2,3,4,5,6]
y = [11,12,13,14,15,16]

# 2. setup the pyplot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Ploting the Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Random Numbers", 
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save and Show your graph(Saved as an Whole fig)
fig.savefig("sample.png")
```

### Making Plot From  Dictionaries

### Bar Graph

```python
laptop_prices = {"Hp Omen 15": 112000, "Lenovo Legin 5": 136000, "Acer Nitro 5": 95000,
                 "Asus Rog Strix G15": 110000, "Asus TUF": 88000}
fig, ax = plt.subplots()
ax.set(title = "My Range of Best Laptops")
ax.set_xticklabels(laptop_prices.keys(), rotation='vertical')
ax.bar(laptop_prices.keys(), laptop_prices.values())
```

```python
fig, ax = plt.subplots()
ax.barh(list(laptop_prices.keys()), list(laptop_prices.values()))
```

### Histogram 

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

# Two Options for subplots

```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each differnt axis
ax1.plot(x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.barh(list(laptop_prices.keys()), list(laptop_prices.values()))
ax4.hist(np.random.randn(1000))
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Ploting
ax[0,0].plot(x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].barh(list(laptop_prices.keys()), list(laptop_prices.values()))
ax[1,1].hist(np.random.randn(1000))
```

## OO method of using Matplotlib

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
ax.set_xlim([45,100])
```

```python
# OO method From Scratch
fig, ((ax1, ax2))= plt.subplots(nrows=2, ncols=1 ,figsize=(10, 10))

# Plot the Data
scatter1 = ax1.scatter(over_50["age"], over_50["chol"], c=over_50["target"])
scatter2 = ax2.scatter(over_50['age'], over_50["thalach"], c=over_50["target"])

# Customise the Plot
ax1.set(title='Heart Diseases and cholestrol Levels',
      xlabel="Age",
      ylabel="Cholesterol")

ax2.set(title='Heart Diseases and Max Heart Rate Levels',
      xlabel="Age",
      ylabel="Max Heart Rate")

# Add a legend 
ax1.legend(*scatter1.legend_elements(), title="Target")
ax2.legend(*scatter2.legend_elements(), title="Target")

# Adding a Horizontal Line
ax1.axhline(over_50["chol"].mean(), linestyle="--")
ax2.axhline(over_50["thalach"].mean(), linestyle="--")

# Adding title to the whole Graphs
fig.suptitle("People Over 50 Years w.r.t Cholestrol and Max Heart Rate", fontsize=20, fontweight="bold")
```

```python
# OO method From Scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the Data
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"], cmap="winter", s= 35)

# Customise the Plot
ax.set(title='Heart Diseases and cholestrol Levels',
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend 
ax.legend(*scatter.legend_elements(), title="Target")

# Adding a Horizontal Line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# OO method From Scratch
fig, ((ax1, ax2))= plt.subplots(nrows=2, ncols=1 ,figsize=(10, 10))

# Plot the Data
scatter1 = ax1.scatter(over_50["age"], over_50["chol"], c=over_50["target"], cmap="winter", s= 35)
scatter2 = ax2.scatter(over_50['age'], over_50["thalach"], c=over_50["target"], cmap="summer", s= 35)

# Customise the Plot
ax1.set(title='Heart Diseases and cholestrol Levels',
      ylabel="Cholesterol")
ax1.set_xticks([])
ax1.set_ylim(200, 500)

ax2.set(title='Heart Diseases and Max Heart Rate Levels',
      xlabel="Age",
      ylabel="Max Heart Rate")
ax2.set_ylim(60, 200)


# Add a legend 
ax1.legend(*scatter1.legend_elements(), title="Target")
ax2.legend(*scatter2.legend_elements(), title="Target")

# Adding a Horizontal Line
ax1.axhline(over_50["chol"].mean(), linestyle="--")
ax2.axhline(over_50["thalach"].mean(), linestyle="--")

# Adding title to the whole Graphs
fig.suptitle("People Over 50 Years w.r.t Cholestrol and Max Heart Rate", fontsize=20, fontweight="bold")


# Saving the Figure
fig.savefig("heart_diseases.png")
```
