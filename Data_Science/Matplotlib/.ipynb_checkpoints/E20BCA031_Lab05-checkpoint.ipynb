```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x1 = [1,2,3,4,5,6,7]
y1 = [1,2,3,4,5,6,7]
x2 = [1, 1.5, 2, 2.5,3 , 3.5]
y2 = [2, 2.5, 3, 2.5, 2,  1.5]
fig, ((ax1, ax2))= plt.subplots(nrows=1, ncols=2 ,figsize=(10, 5))

scatter1 = ax1.plot(x1, y1)
scatter2 = ax2.plot(x2, y2)

ax1.set(title='Draw a Line',
      xlabel="X-axis",
      ylabel="y-axis")

ax2.set(title='Sample Graph',
      xlabel="x-axis",
      ylabel="Y-axis")

fig.suptitle("Assignment Data", fontsize=20, fontweight="bold")
```

```python
# Date,      Open,          High,         Low,         Close
# 10-03-21,  774.25,      776.065002,   769.5,         772.559998
# 10-04-21,  776.030029,  778.710022,   772.890015,    776.429993
# 10-05-21,  779.309998,  782.070007,   775.650024,    776.469971
# 10-06-21,  779,         780.47998,    775.539978,    776.859985
# 10-07-21,  779.659973,  779.659973,   770.75,        775.080017

data = pd.DataFrame({"Date": ["10-03-21", "10-04-21", "10-05-21", "10-06-21", "10-07-21"], 
                  "Open" : [774.25, 776.030029, 779.309998, 779, 779.659973], 
                  "High" : [776.065002, 778.710022, 782.070007, 780.47998, 779.659973], 
                  "Low" : [769.5, 772.890015, 775.650024, 775.539978, 770.75], 
                  "Close" : [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
                  })

data.to_csv("xyz.csv")
```

```python
xyz = pd.read_csv("xyz.csv")
xyz
```

```python
fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(nrows=2, ncols=2 ,figsize=(10, 8))

scatter1 = ax1.bar(xyz["Date"], xyz["Open"])
scatter2 = ax2.plot(xyz["Date"], xyz["High"])
scatter3 = ax3.scatter(xyz["Low"],xyz["Date"])
scatter4 = ax4.bar(xyz["Date"], xyz["Close"])
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.8)

ax1.set(title='W.R.T Open',
      xlabel="X-axis",
      ylabel="y-axis")

ax2.set(title='W.R.T High',
      xlabel="x-axis",
      ylabel="Y-axis")


ax3.set(title='W.R.T Low',
      xlabel="x-axis",
      ylabel="Y-axis")

ax4.set(title='W.R.T Close',
      xlabel="x-axis",
      ylabel="Y-axis")

fig.suptitle("Assignment Data", fontsize=20, fontweight="bold")
```
