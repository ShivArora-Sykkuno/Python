```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
X = df.iloc[:, 3:13].values
Y = df.iloc[:, 13].values
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import LabelEncoder

label_1 = LabelEncoder()
X[:, 1] = label_1.fit_transform(X[:, 1])

label_2 = LabelEncoder()
X[:, 2] = label_2.fit_transform(X[:, 2])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=101) 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
def build_network(opti):
    classifier = Sequential()
    classifier.add(Dense(units=5, kernel_initializer="random_uniform", activation = "relu", input_dim=10))
    classifier.add(Dense(units=5, kernel_initializer="random_uniform", activation = "relu"))
    classifier.add(Dense(units=1, kernel_initializer="random_uniform", activation = "sigmoid"))
    classifier.compile(optimizer=opti, loss="binary_crossentropy", metrics = ["accuracy"])
    return classifier
```

```python
classifier = KerasClassifier(build_fn=build_network)

parms = {"batch_size" : [25, 30],
         "epochs" : [100, 500],
         "opti" : ["adam", "rmsprop"]
        }
grid_search = GridSearchCV(estimator=classifier, param_grid=parms, scoring="accuracy", cv=10, verbose=1, n_jobs=-1)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_
```
