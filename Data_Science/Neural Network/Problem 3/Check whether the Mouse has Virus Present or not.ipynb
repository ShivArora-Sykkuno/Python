```python
X = df.drop("Virus Present", axis =1)
Y = df["Virus Present"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, Y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
def build_network(opti):
    from keras.models import Sequential
    from keras.layers import Dense
    
    classifier = Sequential()
    classifier.add(Dense(units=2, kernel_initializer="random_uniform", activation = "relu", input_dim=2))
    classifier.add(Dense(units=2, kernel_initializer="random_uniform", activation = "relu"))
    classifier.add(Dense(units=1, kernel_initializer="random_uniform", activation = "sigmoid"))
    classifier.compile(optimizer=opti, loss="binary_crossentropy", metrics = ["accuracy"])
    return classifier
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

classifier = KerasClassifier(build_fn=build_network)

parms = {"batch_size" : [25, 30],
         "epochs" : [500, 1000],
         "opti" : ["adam", "rmsprop"]
        }
grid_search = GridSearchCV(estimator=classifier, param_grid=parms, scoring="accuracy", cv=10, verbose=1, n_jobs=-1)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_
```
