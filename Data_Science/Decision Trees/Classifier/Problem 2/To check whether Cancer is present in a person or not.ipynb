```python
X = df.drop("Cancer Present", axis=1)
Y = df["Cancer Present"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
```

```python
grid = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 3, 4, 5, 6], 
    "splitter" : ["best", "random"]
       }
```

```python
from sklearn.model_selection import GridSearchCV

grid_clf = GridSearchCV(clf, grid, cv=10, n_jobs=-1)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.best_params_
```

```python
y_preds = grid_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix

confusion_matrix(y_test, y_preds)
```

```python
print(f"{accuracy_score(y_test, y_preds)*100:.1f}%")
```

## On Training DataSet

```python
plot_confusion_matrix(grid_clf, X_train, y_train)
```

## On testing DataSet

```python
plot_confusion_matrix(grid_clf, X_test, y_test)
```
