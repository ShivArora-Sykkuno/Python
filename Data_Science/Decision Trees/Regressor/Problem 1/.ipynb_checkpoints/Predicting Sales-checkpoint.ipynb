```python
X = df.drop("sales", axis=1)
Y = df["sales"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor

dclf =  DecisionTreeRegressor()
```

```python
grid = {
    "criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter": ["best", "random"],
    "max_depth" : [None, 1, 2, 3, 4, 5, 6]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_dclf = GridSearchCV(dclf, grid, cv=10, n_jobs=-1)
grid_dclf.fit(X, Y)
```

```python
grid_dclf.best_params_
```

```python
y_preds = grid_dclf.predict(X_test)
```

```python
from sklearn.metrics import plot_roc_curve, mean_absolute_error

print(mean_absolute_error(y_preds, y_test))
```
